```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
values, counts = np.unique(x, return_counts=True)
most_freq_value = values[np.argmax(counts)]
indices_x = np.where(x == most_freq_value)

print(most_freq_value) 
print(indices_x)
```

```python
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
sum_all = np.sum(gfg)
sum_row = np.sum(gfg, axis=1)
sum_col = np.sum(gfg, axis=0)
print(sum_all)
print(sum_row) 
print(sum_col)
```

```python
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
diag_sum = np.trace(n_array)
eig_vals, eig_vecs = np.linalg.eig(n_array)
inverse_matrix = np.linalg.inv(n_array)
determinant = np.linalg.det(n_array)
print("Diagonal sum:", diag_sum)
print("Eigenvalues:", eig_vals)
print("Eigenvectors:\n", eig_vecs)
print("Inverse:\n", inverse_matrix)
print("Determinant:", determinant)
```

```python
p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])
mul1 = np.dot(p, q)
cov1 = np.cov(p.T, q.T)
print("Multiplication (i):\n", mul1)
print("Covariance (i):\n", cov1)
```

```python
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
inner_prod = np.inner(x, y)
outer_prod = np.outer(x, y)
cartesian_prod = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1, 2)
print("Inner:\n", inner_prod)
print("\nOuter:\n", outer_prod)
print("\nCartesian:\n", cartesian_prod)
```

```python
#Q2
array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_val = np.abs(array)
flat = array.flatten()
p_flat = np.percentile(flat, [25, 50, 75])
p_col = np.percentile(array, [25, 50, 75], axis=0)
p_row = np.percentile(array, [25, 50, 75], axis=1)
mean_flat, median_flat, std_flat = np.mean(flat), np.median(flat), np.std(flat)
mean_col, median_col, std_col = np.mean(array, axis=0), np.median(array, axis=0), np.std(array, axis=0)
mean_row, median_row, std_row = np.mean(array, axis=1), np.median(array, axis=1), np.std(array, axis=1)
print("Absolute:\n", abs_val)
print("Percentiles flat:", p_flat)
print("Percentiles col:\n", p_col)
print("Percentiles row:\n", p_row)
print("Mean flat:", mean_flat, "Median flat:", median_flat, "Std flat:", std_flat)
```

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_val = np.floor(a)
ceil_val = np.ceil(a)
trunc_val = np.trunc(a)
round_val = np.round(a)
print("Floor:", floor_val)
print("Ceiling:", ceil_val)
print("Truncate:", trunc_val)
print("Round:", round_val)
```

```python
#q3
arr = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_arr = np.sort(arr)
indices_sorted = np.argsort(arr)
smallest4 = np.sort(arr)[:4]
largest5 = np.sort(arr)[-5:]
print("Sorted:", sorted_arr)
print("Indices:", indices_sorted)
print("4 smallest:", smallest4)
print("5 largest:", largest5)
```

```python
arr2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = arr2[arr2 == np.floor(arr2)]
float_elements = arr2[arr2 != np.floor(arr2)]
print("Integers:", integer_elements)
print("Floats:", float_elements)
```
